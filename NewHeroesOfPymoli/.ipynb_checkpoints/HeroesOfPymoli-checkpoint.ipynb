* Calculate the numbers and percentages by age group

* Optional: round the percentage column to two decimal points

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
gamers = pd.read_csv(file_to_load)
```

```python
bins = [0,9,14,19,24,29,34,39,99]

labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "+40"]
```

```python
gamers["Age Summary"] = pd.cut( gamers["Age"], bins, labels =labels)
```

```python
pd.cut(gamers["Age"], bins)
```

```python
pd.cut(gamers["Age"], bins, labels= labels)
gamers["Age Summary"] 
gamers.dtypes
AgeInfo = gamers["Age Summary"] 
counts = AgeInfo.value_counts()
percent = AgeInfo.value_counts(normalize=True).mul(100).round(2).astype(str) + "%"
pd.DataFrame({'Counts': counts, 'Percent': percent})
```

## Top Spenders ok

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
# Read Purchasing File and store into Pandas data frame
gamers = pd.read_csv(file_to_load)
gamers['Purchase Count'] = gamers.groupby("SN")["SN"].transform('count')

#gamers
gamers[["SN", "Price", "Purchase Count"]]
gamers[["Price","Count"]] = gamers[["Price", "Purchase Count"]].astype(float)
gamers ["Total Purchase"]= gamers["Price"] * gamers["Count"]

gamers.sort_values("Total Purchase", ascending= False)

gamers ["Avg Purchase Price"]= gamers["Total Purchase"] / gamers["Count"]
gamers.sort_values("Avg Purchase Price", ascending= False)
TopSpenders = gamers.groupby("SN")
gamers.set_index("SN")
gamers.sort_values(by = "Total Purchase", ascending = False )
TopItems = gamers.groupby("SN")
gamers.set_index("SN")
gamers.sort_values(by = "Total Purchase", ascending = False )

```

## Most Popular Items ok

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
file_to_load = "Resources/purchase_data.csv"
# Read Purchasing File and store into Pandas data frame
gamers = pd.read_csv(file_to_load)

#Add a column for Purchase Count
gamers['Purchase Count'] = gamers.groupby("Item ID")["Item Name"].transform('count')
```

```python
gamers ["Total Purchase"]= gamers["Price"] * gamers["Purchase Count"].astype (float)
```

```python
#Retrieve the Item ID, Item Name, and Item Price columns
MostPopularItems= gamers[["Item ID", "Item Name", "Price", "Purchase Count", "Total Purchase" ]].head()
MostPopularItems.sort_values("Purchase Count", ascending = False)
```
